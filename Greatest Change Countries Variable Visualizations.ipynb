```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
world_happy_by_year = pd.read_csv("Resources/yearly_country_data.csv")
world_happy_by_year
```

```python
afghanistan = world_happy_by_year.loc[world_happy_by_year['Country'] == 'Afghanistan']
afghanistan
```

```python
Liberia = world_happy_by_year.loc[world_happy_by_year['Country'] == 'Liberia']
Liberia
```

```python
plt.plot(afghanistan['Year'], afghanistan['Family'], marker='s', color='pink', label='Family')
plt.xlabel('Year')
plt.ylabel('Family')
plt.title('Change in Family in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/family_afghanistan.png")
```

```python
plt.plot(afghanistan['Year'], afghanistan['Freedom'], marker='*', color='red', label='Freedom')
plt.xlabel('Year')
plt.ylabel('Freedom')
plt.title('Change in Sense of Freedom in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/freedom_afghanistan.png")
```

```python
plt.plot(afghanistan['Year'], afghanistan['Generous'], marker='p', color='purple', label='Generosity')
plt.xlabel('Year')
plt.ylabel('Generosity')
plt.title('Change in Generosity in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/generosity_afghanistan.png")
```

```python
plt.plot(afghanistan['Year'], afghanistan['Trust '], marker='4', color='orange', label='Trust')
plt.xlabel('Year')
plt.ylabel('Mistrust')
plt.title('Change in Mistrust of Government and Business in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/trust_afghanistan.png")
```

```python
plt.plot(afghanistan['Year'], afghanistan['Health'], marker='P', color='blue', label='Health')
plt.xlabel('Year')
plt.ylabel('Health')
plt.title('Change in Health in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/health_afghanistan.png")
```

```python
plt.plot(afghanistan['Year'], afghanistan['Economy'], marker='^', color='black', label='Economy')
plt.xlabel('Year')
plt.ylabel('Economy')
plt.title('Change in Economy in Afghanistan')
plt.legend()
plt.savefig("Analysis_Figures_Images/economy_afghanistan.png")
```
