```python
import pandas as pd
import numpy as np
```

```python
gini_df = pd.read_csv("Resources/Gini_Index_Data.csv")
```

```python
gini_df.head()
```

```python
gini_df = gini_df[["Country Name", "2015[YR2015]", "2016[YR2016]", "2017[YR2017]", "2018[YR2018]", "2019[YR2019]"]]
gini_df.head()
len(gini_df.index)
```

```python
gini_df.head()
```

```python
null_values = gini_df.loc[(gini_df["2015[YR2015]"] == "..") & (gini_df["2016[YR2016]"] == "..") & (gini_df["2017[YR2017]"] == "..") & (gini_df["2018[YR2018]"] == "..") & (gini_df["2019[YR2019]"] == "..")]
null_index = null_values.index

new_gini_df = gini_df.replace("..", np.nan)
new_gini_df.head()
```

```python
gini2_df = new_gini_df.drop(null_index)
gini2_df.head()
```

```python
gini2_df['2015[YR2015]'] = gini2_df['2015[YR2015]'].astype(float)
gini2_df['2016[YR2016]'] = gini2_df['2016[YR2016]'].astype(float)
gini2_df['2017[YR2017]'] = gini2_df['2017[YR2017]'].astype(float)
gini2_df['2018[YR2018]'] = gini2_df['2018[YR2018]'].astype(float)
gini2_df['2019[YR2019]'] = gini2_df['2019[YR2019]'].astype(float)
```

```python
gini3_df = gini2_df.set_index("Country Name", inplace=True)
```

```python
gini2_df.head()

gini2_df = gini2_df.drop(["level_0", "index"], axis=1)
```

```python
gini2_df["Gini Index Five Year Average"] = gini2_df.mean(numeric_only=True, axis=1)
gini2_df.head()
```

```python
gini2_df.to_csv("Gini_Index_df.csv")
```
