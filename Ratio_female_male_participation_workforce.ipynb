```python
import pandas as pd
import numpy as np


file_path = 'gender_ratio_workforce_data/251d2418-5578-4a67-bb41-4ea24ad274bf_Data.csv'
```

```python
ratio_df = pd.read_csv(file_path)
ratio_df.head()
```

```python
ratio_df = ratio_df[['Country Name', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']]
ratio_df.head()
len(ratio_df.index)
```

```python
ratio_df.head()
null_values = ratio_df.loc[(ratio_df['2015 [YR2015]'] == '..') & (ratio_df['2016 [YR2016]'] == '..') & (ratio_df['2017 [YR2017]'] == '..') & (ratio_df['2018 [YR2018]'] == '..') & (ratio_df['2019 [YR2019]'] == '..')]
null_index = null_values.index
```

```python
ratio_df = ratio_df.drop(null_index)
ratio_df.head()
```

```python
ratio_df['2015 [YR2015]'] = ratio_df['2015 [YR2015]'].astype(float)
ratio_df['2016 [YR2016]'] = ratio_df['2016 [YR2016]'].astype(float)
ratio_df['2017 [YR2017]'] = ratio_df['2017 [YR2017]'].astype(float)
ratio_df['2018 [YR2018]'] = ratio_df['2018 [YR2018]'].astype(float)
ratio_df['2019 [YR2019]'] = ratio_df['2019 [YR2019]'].astype(float)
```

```python
ratio_df['Workforce Ratio Five Year Average'] = ratio_df[['2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']].mean(axis=1)
ratio_df.head()
```

```python
ratio_df['Workforce Ratio Five Year Median'] = ratio_df[['2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']].median(axis=1)
ratio_df['Workforce Ratio Five Year Max'] = ratio_df[['2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']].max(axis=1)
ratio_df['Workforce Ratio Five Year Min'] = ratio_df[['2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']].min(axis=1)
ratio_df['Workforce Ratio Five Year STD'] = ratio_df[['2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]']].std(axis=1, numeric_only=True)
```

```python
ratio_df.head()
```

```python
ratio_df.rename(columns={
    '2015 [YR2015]': 'Workforce Ratio 2015',
    '2016 [YR2016]': 'Workforce Ratio 2016',
    '2017 [YR2017]': 'Workforce Ratio 2017',
    '2018 [YR2018]': 'Workforce Ratio 2018',
    '2019 [YR2019]': 'Workforce Ratio 2019'
})
```
