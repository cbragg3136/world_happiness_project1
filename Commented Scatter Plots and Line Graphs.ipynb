```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#read in resources
world_happy_year = pd.read_csv("Resources/happy_by_year.csv")
all_variables_ave = pd.read_csv("Resources/all_variables_avg.csv")
```

```python
#create new dataframe for line graphs by adding Gini Index data and workforce ratio data (no data for 2019 Gini Index)
world_happy_year_ave = world_happy_year[["Year", "Happy", "LogGDP", 
                                        "Family", "Health", "Freedom", 
                                         "Generous", "Trust"]]
world_year_complete = world_happy_year_ave.assign(Workforce_Ratio_mean = [70.64387477, 70.832537, 70.96893882, 71.19846345, 71.23556771], 
                                                  GINI_Index_mean = [36.7975, 36.35789, 36.10299, 39.95172, 'NaN'])
world_year_complete.to_csv("Resources/world_year_complete.csv")
world_year_complete
```

```python
#define linear regression function for scatter plots
def linreg(x_values, y_values, title, coordinates):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,coordinates,fontsize=15,color="red")
    plt.xlabel(title)
    plt.ylabel('Happy')
    print(f"The r squared value is: {rvalue}")
    plt.show() 
```

```python
#begin scatter plots and linear regressions for country 5 year average of happiness score vs. 5 year average of each variable
fig = all_variables_ave.plot(kind='scatter', x='Economy', y='Happy').get_figure()
plt.title("Happiness Score vs Economy")
plt.grid()
x_values = all_variables_ave["Economy"]
y_values = all_variables_ave["Happy"]
linreg(all_variables_ave["Economy"],all_variables_ave["Happy"], "Economy", (7, 7))
fig.savefig('Analysis_Figures_Images/happy_vs_economy.png')
```
