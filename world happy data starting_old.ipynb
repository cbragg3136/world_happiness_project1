```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
#Load resources files from world happiness dataset
wor_hap_2015 = pd.read_csv("Resources/2015.csv")
wor_hap_2016 = pd.read_csv("Resources/2016.csv")    
wor_hap_2017 = pd.read_csv("Resources/2017.csv")
wor_hap_2018 = pd.read_csv("Resources/2018.csv")
wor_hap_2019 = pd.read_csv("Resources/2019.csv")
```

```python
#wor_hap_2015.head()
```

```python
#2015 year added in to 2015 Happy Data
year15 = []
year15 = [2015]*len(wor_hap_2015)

wor_hap_2015['year'] = year15

clean_hap_2015 = wor_hap_2015

#cleaning names/removing columns
del clean_hap_2015['Standard Error']
del clean_hap_2015['Region']
del clean_hap_2015['Dystopia Residual']

clean_hap_2015.head()
```

```python
#wor_hap_2016.head()
```

```python
#2016 year added in to 2016 Happy Data
year16 = []
year16 = [2016]*len(wor_hap_2016)
  
wor_hap_2016['year'] = year16

clean_hap_2016 = wor_hap_2016

#cleaning names/removing columns

del clean_hap_2016['Lower Confidence Interval']
del clean_hap_2016['Upper Confidence Interval']
del clean_hap_2016['Region']
del clean_hap_2016['Dystopia Residual']

clean_hap_2016.head()
```

```python
#2017 year added in to 2017 Happy Data
year17 = []
year17 = [2017]*len(wor_hap_2017)
  
wor_hap_2017['year'] = year17

clean_hap_2017 = wor_hap_2017

#cleaning names/removing columns

del clean_hap_2017['Whisker.high']
del clean_hap_2017['Whisker.low']
del clean_hap_2017['Dystopia.Residual']

clean_hap_2017 = clean_hap_2017.rename(columns = {'Happiness.Rank': "Happiness Rank", 'Happiness.Score':'Happiness Score', 'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)'})
clean_hap_2017 = clean_hap_2017.rename(columns = {'Health..Life.Expectancy.':'Health (Life Expectancy)', 'Trust..Government.Corruption.':'Trust (Government Corruption)'})

#fix the order to be consistent with other years
clean_hap_2017 = clean_hap_2017[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','year']]

clean_hap_2017.head()
```

```python
#2018 year added in to 2018 Happy Data
year18 = []
year18 = [2018]*len(wor_hap_2018)
  
wor_hap_2018['year'] = year18

clean_hap_2018 = wor_hap_2018

#cleaning names/removing columns
clean_hap_2018 = clean_hap_2018.rename(columns = {'Overall rank': 'Happiness Rank','Country or region':'Country','Score':'Happiness Score' })
clean_hap_2018 = clean_hap_2018.rename(columns = {'GDP per capita':'Economy (GDP per Capita)', 'Social support':'Family', 'Healthy life expectancy':'Health (Life Expectancy)'})
clean_hap_2018 = clean_hap_2018.rename(columns = {'Freedom to make life choices':'Freedom', 'Perceptions of corruption':'Trust (Government Corruption)'})

#fix the order to be consistent with other years
clean_hap_2018 = clean_hap_2018[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','year']]

clean_hap_2018.head()
```

```python
#2019 year added in to 2019 Happy Data
year19 = []
year19 = [2019]*len(wor_hap_2019)
  
wor_hap_2019['year'] = year19

clean_hap_2019 = wor_hap_2019

#cleaning names/removing columns
clean_hap_2019 = clean_hap_2019.rename(columns = {'Overall rank': 'Happiness Rank','Country or region':'Country','Score':'Happiness Score' })
clean_hap_2019 = clean_hap_2019.rename(columns = {'GDP per capita':'Economy (GDP per Capita)', 'Social support':'Family', 'Healthy life expectancy':'Health (Life Expectancy)'})
clean_hap_2019 = clean_hap_2019.rename(columns = {'Freedom to make life choices':'Freedom', 'Perceptions of corruption':'Trust (Government Corruption)'})

#fix the order to be consistent with other years
clean_hap_2019 = clean_hap_2019[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','year']]


clean_hap_2019.head()
```

```python
# one merge method - #merge the data files together 

merged_hap_15_16 = pd.merge(clean_hap_2015, clean_hap_2016,  on='Country', how='left')
merged_hap_15_17 = pd.merge(merged_hap_15_16, clean_hap_2017, on='Country', how='left')
merged_hap_15_18 = pd.merge(merged_hap_15_17, clean_hap_2018, on='Country', how='left')
merged_hap_15_19 = pd.merge(merged_hap_15_18, clean_hap_2019, on='Country', how='left')

merged_hap_15_19.head()
```

```python
#stacking all data tables together via concatenation (by country, by year)
#https://www.freecodecamp.org/news/how-to-combine-multiple-csv-files-with-8-lines-of-code-265183e0854/
stack_15_16 = pd.concat([clean_hap_2015, clean_hap_2016, clean_hap_2017, clean_hap_2018, clean_hap_2019], axis = 0)

stack_15_16
```

```python
#building a table of summary statistics by year - all countries

basic_data_year = stack_15_16.groupby('year').agg([np.mean, np.median, np.min, np.max, np.std])
basic_data_year
```

```python
#export summary table of happiness data by year
```

```python
#building a table of summary statistics by country

basic_data_country = stack_15_16.groupby('Country').agg([np.mean, np.median, np.min, np.max, np.std])
basic_data_country
```

```python
#export summary table of happiness data by country
wor_happy_data_by_country = basic_data_country
wor_happy_data_by_country.to_csv(r'happy_by_country.csv')
```

```python
# building a correlation table
# Happiness Rank	Happiness Score	Economy (GDP per Capita)	Family	Health (Life Expectancy)	Freedom	Trust (Government Corruption)	Generosity
happy_score = basic_data_country['Happiness Score', 'mean']
economy_score = basic_data_country['Economy (GDP per Capita)', 'mean']
family_score = basic_data_country['Family', 'mean']
health_score = basic_data_country['Health (Life Expectancy)', 'mean']
freedom_score = basic_data_country['Freedom', 'mean']
trust_score = basic_data_country['Trust (Government Corruption)', 'mean']
generous_score = basic_data_country['Generosity', 'mean']

# make a new dictionary

correl_table = { 
    'Happy': happy_score, 
    "Economy": economy_score, 
    "Family": family_score,
    "Health": health_score,
    "Freedom": freedom_score,
    "Trust": trust_score,
    "Generous": generous_score
    }

correl_table = pd.DataFrame(correl_table)
correl_table
```
